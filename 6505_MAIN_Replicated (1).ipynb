## H1: Self-concept dimensions (i.e., physical, social, temperamental, educational, moral, and intellectual) increases the impact of anxiety.

To determine the relationship between self-concept and anxiety, the study tested six parameters to predict their impact on
anxiety. The tested parameters were: Physical, Social, Temperamental, Educational, Moral and intellectual. The correlation
coefficient, R, was calculated and used to predict the relationship between self-concept and anxiety

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm


# Independent variables (predictors)
X = df[['Intellectual', 'Moral', 'Educational', 'Social', 'Physical', 'Temperamental']]

# Dependent variable
y = df['TOTAL_BAI']

# Add constant for statsmodels
X_with_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_with_const).fit()

# Print the model summary
print(model.summary())

# Extract R, R^2, Adjusted R^2, and Std. Error of the Estimate
r_square = model.rsquared
adjusted_r_square = model.rsquared_adj

# Number of observations and predictors
n = len(y)
p = X.shape[1]
# Calculate Residual Sum of Squares (RSS)
rss = np.sum((y - model.fittedvalues) ** 2)

# Calculate Std. Error
std_error_rss = np.sqrt(rss / (n - p - 1))
r = np.sqrt(r_square)

print("==============================================================================")

print(f"\nR: {r}")
print(f"R^2: {r_square}")
print(f"Adjusted R^2: {adjusted_r_square}")
print(f"Std. Error of the Estimate:{std_error_rss}")
```

A multiple regression analysis was conducted to examine the relationship between several predictor variables (Intellectual, Moral, Educational, Social, Physical, and Temperamental) and the outcome variable, TOTAL_BAI. The model was statistically significant, *F*(6, 204) = 9.09, *p* < .001, accounting for 21.1% of the variance in TOTAL_BAI.

Of the predictor variables, only Physical and Temperamental had a significant negative relationship with TOTAL_BAI, *β* = -.84, *p* < .001, and *β* = -.67, *p* < .05, respectively. The remaining predictors were not significantly related to the outcome.

```python
# Regression ANOVA without SCaling

# Define predictors and dependent variable
X = df[["Physical", "Social", "Temperamental", "Educational", "Moral", "Intellectual"]]
y = df["TOTAL_BAI"]

# Add constant to the model
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Get Regression Sum of Squares (SSR), Residual Sum of Squares (SSE), and Total Sum of Squares (SST)
sst = ((y - y.mean()) ** 2).sum()  # Total sum of squares
ssr = model.ess  # Regression sum of squares
sse = model.ssr  # Residual sum of squares

# Degrees of freedom
df_model = model.df_model  # Number of predictors
df_resid = model.df_resid  # Residual degrees of freedom
df_total = len(y) - 1  # Total degrees of freedom

# Mean Squares
msr = ssr / df_model  # Mean square for regression
mse = sse / df_resid  # Mean square for residuals

# F-statistic
f_stat = msr / mse

# p-value for F-statistic
p_value = model.f_pvalue

# Display ANOVA table
anova_table = pd.DataFrame({
    "Source": ["Regression", "Residual", "Total"],
    "Sum of Squares": [ssr, sse, sst],
    "df": [df_model, df_resid, df_total],
    "Mean Square": [msr, mse, None],
    "F": [f_stat, None, None],
    "Sig.": [p_value, None, None]
})

print("ANOVA Table:")
print(anova_table.round(3))
```

The effect of the six self-concept dimensions (Intellectual, Educational, Moral, Social, Physical, and Temperamental) on anxiety (BAI). The results indicated a significant effect of the self-concept dimensions on anxiety, F(6, 204) = 9.088, p < .001.

## 3.1 Estimated model Coefficients

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler



# Define predictors and dependent variable
X = df[['Physical', 'Social', 'Temperamental', 'Educational', 'Moral', 'Intellectual']]
y = df['TOTAL_BAI']

# Add constant to the model
X_with_const = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X_with_const).fit()

# Standardize the independent variables (Z-score normalization)
scaler = StandardScaler()
X_standardized = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_standardized_with_const = sm.add_constant(X_standardized)

# Fit the OLS model for standardized coefficients
model_standardized = sm.OLS(y, X_standardized_with_const).fit()

# Extract results for unstandardized and standardized coefficients
coefficients = pd.DataFrame({
    'Unstandardised Coefficients (B)': model.params,
    'Standardised Coefficients (Beta)': model_standardized.params[1:],  # Skip constant for standardized
    'Std. Error': model.bse,
    't': model.tvalues,
    'Sig.': model.pvalues
})

# Print the table
print(coefficients.round(3))
```

An unstandardized coefficient indicates how much the dependent variable varies with an independent variable.

Among the predictors, Physical and Temperamental were statistically significant predictors of the dependent variable (p < .05 for both), demonstrating strong negative relationships. The predictors Educational, Intellectual, and Social had effects that were not statistically significant (p > .05 for all), indicating weaker or inconsistent relationships. The constant term provided a reliable baseline value for the dependent variable when all predictors were zero. These findings emphasize the importance of Physical and Temperamental in predicting the dependent variable, while other predictors may require further investigation or could be excluded from the model due to their lack of significance.

##  H2: There is a significant mean difference between the physical, social, Temperamental, Educational, Intellectual, and Moral self-concepts of males and females. - Group Statistics -

```python
import pandas as pd
import numpy as np

# Group the DataFrame by 'Gender'
grouped_df = df.groupby('Gender')

# List of columns to summarize
columns_to_summarize = ['Physical', 'Social', 'Temperamental', 'Educational', 'Intellectual', 'Moral', 'TOTAL_BAI']

# Initialize a list to store the results
results = []

# Loop through the columns and calculate statistics for each group
for col in columns_to_summarize:
    for gender, group in grouped_df:
        n = group[col].count()
        mean = group[col].mean()
        std_dev = group[col].std()
        std_error = std_dev / np.sqrt(n)
        results.append([col, gender, n, mean, std_dev, std_error])

# Create a DataFrame for the results
summary_table = pd.DataFrame(
    results,
    columns=['Variable', 'Gender', 'N', 'Mean', 'Std. Deviation', 'Std. Error Mean']
)

# Format the table while preserving the specified order
summary_table['Variable'] = pd.Categorical(summary_table['Variable'], categories=columns_to_summarize, ordered=True)
summary_table_sorted = summary_table.sort_values('Variable')

# Pivot the sorted DataFrame
summary_table_formatted = summary_table_sorted.pivot_table(
    index='Variable',
    columns='Gender',
    values=['N', 'Mean', 'Std. Deviation', 'Std. Error Mean']
)

# Print the formatted table
print("Table 3.4: Group Statistics")
print(summary_table_formatted.to_string())
```

A **descriptive analysis** was conducted to compare the self-concept dimensions (physical, social, temperamental, educational, intellectual, and moral) between males and females. The results are as follows:

3. **Temperamental Self-Concept**:

4. **Educational Self-Concept**:
   - Females (M = 27.20, SD = 4.12) scored higher than males (M = 26.57, SD = 4.47). The standard error was 0.40 for females and 0.44 for males, indicating consistent variability.

5. **Intellectual Self-Concept**:

## H3: There is a significant mean difference between the anxiety scores of males and females.

## Independent samples test

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Load your data

# Map Gender values to 0 and 1 (assuming 1 for male, 2 for female)
df['Gender_num'] = df['Gender'].map({'M': 1, 'F': 2})

# Define the columns to test
variables = ['Physical', 'Social', 'Temperamental', 'Educational', 'Moral', 'Intellectual', 'TOTAL_BAI']

# Function to perform independent samples t-test
def perform_t_tests(df, variables):
    results = []
    for var in variables:
        group1 = df[df['Gender_num'] == 1][var]
        group2 = df[df['Gender_num'] == 2][var]

        # Levene's Test for Equality of Variances
        levene_stat, levene_p = stats.levene(group1, group2)

        # T-test for Equality of Means
        t_stat_eq_var, t_p_eq_var = stats.ttest_ind(group1, group2, equal_var=True)
        t_stat_uneq_var, t_p_uneq_var = stats.ttest_ind(group1, group2, equal_var=False)

        # Calculate mean difference and confidence intervals
        mean_diff = group1.mean() - group2.mean()

        # Equal variances assumed
        std_err_diff_eq_var = np.sqrt((group1.var() / len(group1)) + (group2.var() / len(group2)))
        df_eq_var = len(group1) + len(group2) - 2
        ci_lower_eq_var = mean_diff - 1.96 * std_err_diff_eq_var
        ci_upper_eq_var = mean_diff + 1.96 * std_err_diff_eq_var

        # Equal variances not assumed
        se_diff_uneq_var = np.sqrt((group1.var()/len(group1)) + (group2.var()/len(group2)))
        df_uneq_var = ((group1.var()/len(group1)) + (group2.var()/len(group2)))**2 / ((group1.var()**2/(len(group1)**2*(len(group1)-1))) + (group2.var()**2/(len(group2)**2*(len(group2)-1))))
        ci_lower_uneq_var = mean_diff - 1.96 * se_diff_uneq_var
        ci_upper_uneq_var = mean_diff + 1.96 * se_diff_uneq_var

        results.append({
            'Variable': var,
            'Levene\'s Test F': levene_stat,
            'Levene\'s Test p-value': levene_p,
            't-statistic (Equal Var)': t_stat_eq_var,
            'df (Equal Var)': df_eq_var,
            't-test p-value (Equal Var)': t_p_eq_var,
            'Mean Difference': mean_diff,
            'Std. Error Diff (Equal Var)': std_err_diff_eq_var,
            '95% CI Lower (Equal Var)': ci_lower_eq_var,
            '95% CI Upper (Equal Var)': ci_upper_eq_var,
            't-statistic (Unequal Var)': t_stat_uneq_var,
            'df (Unequal Var)': df_uneq_var,
            't-test p-value (Unequal Var)': t_p_uneq_var,
            'Std. Error Diff (Unequal Var)': se_diff_uneq_var,
            '95% CI Lower (Unequal Var)': ci_lower_uneq_var,
            '95% CI Upper (Unequal Var)': ci_upper_uneq_var
        })
    return pd.DataFrame(results)

# Perform the t-tests
results_df = perform_t_tests(df, variables)

# Display the results
print(results_df)
```
